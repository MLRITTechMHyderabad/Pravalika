```python
import numpy as np

points = np.array([
    [10, 25, 30, 22, 12, 18, 26, 32, 24, 29],
    [20, 15, 12, 28, 24, 26, 30, 18, 22, 25],
    [35, 30, 32, 40, 28, 34, 29, 31, 26, 37],
    [12, 18, 20, 15, 22, 14, 19, 21, 23, 17],
    [28, 26, 27, 25, 30, 29, 35, 32, 31, 38]
])

avg_points = points.mean(axis=1)
total_points = points.sum(axis=1)
best_player = np.argmax(total_points)
worst_player = np.argmin(total_points)
games_above_30 = {f"Player {i+1}": np.where(points[i] > 30)[0]+1 for i in range(5)}
sorted_players = sorted(enumerate(total_points), key=lambda x: x[1], reverse=True)

print(avg_points)
print(f"Best player: Player {best_player+1} ({total_points[best_player]} points)")
print(f"Worst player: Player {worst_player+1} ({total_points[worst_player]} points)")
for player, games in games_above_30.items():
    print(f"{player}: Games {games}")
for idx, total in sorted_players:
    print(f"{idx+1}. Player {idx+1} - {total} points")
```
